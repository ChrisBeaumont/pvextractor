```python
url = 'http://files.figshare.com/1213488/iras05358_12co21.fits'
```

```python
from astropy.utils.data import get_readable_fileobj
from astropy.io import fits
from astropy import wcs
```

```python
with get_readable_fileobj(url, cache=True) as f:
    fitsfile = fits.open(f)
    data = fitsfile[0].data
    header = fitsfile[0].header
```

```python
w = wcs.WCS(header)
```

```python
data.shape
```

```python
import aplpy
```

```python
F = aplpy.FITSFigure(fitsfile,slices=[500])
F.show_grayscale()
```

```python
from pvextractor import pvdiagram,pvwcs_from_header
```

```python
endpoints = [(0,0),(20,20)]
pv,npts = pvdiagram(data, endpoints=endpoints)
```

```python
pv.shape
```

```python
w2 = pvwcs_from_header(header)
h2 = w2.to_header()
hdu = fits.PrimaryHDU(data=pv,header=h2)
```

```python
F2 = aplpy.FITSFigure(hdu)
F2.show_grayscale(aspect=pv.shape[1]/float(pv.shape[0]), stretch='arcsinh')
```

```python
# a hacktacular way to overplot a line on the "flat" image
endpoints_wcs = F._wcs.wcs_pix2world([[x,y,0] for x,y in endpoints], 0)
```

```python
F = aplpy.FITSFigure(fitsfile,slices=[500])
F.show_grayscale()
F.show_lines([endpoints_wcs[:,:2].T],color='r')
```

Now for a weirder case, with more endpoints

```python
endpoints = [(8,8),(8,15),(15,15),(15,8),(3,5)]
endpoints_wcs = F._wcs.wcs_pix2world([[x,y,0] for x,y in endpoints], 0)
F = aplpy.FITSFigure(fitsfile,slices=[500])
F.show_grayscale()
F.show_lines([endpoints_wcs[:,:2].T],color='r')
```

```python
pv,npts = pvdiagram(data, endpoints=endpoints)
w2 = pvwcs_from_header(header)
h2 = w2.to_header()
hdu = fits.PrimaryHDU(data=pv,header=h2)
```

```python
F2 = aplpy.FITSFigure(hdu)
F2.show_grayscale(aspect=pv.shape[1]/float(pv.shape[0]), stretch='arcsinh')
```
